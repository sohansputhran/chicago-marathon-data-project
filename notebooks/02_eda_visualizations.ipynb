```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load cleaned data (update path if needed)
df = pd.read_csv('../data/cleaned_marathon_2023.csv')
df.head()
```

```python
df.head(20)
```

```python
# Convert time columns to timedelta
# This is necessary to handle the time format correctly
df['time'] = pd.to_timedelta(df['time'], errors='coerce')
df['diff'] = pd.to_timedelta(df['diff'], errors='coerce')
df['min/km'] = pd.to_timedelta(df['min/km'], errors='coerce')
df['min/mile'] = pd.to_timedelta(df['min/mile'], errors='coerce')
```

```python
# Convert time to minutes for easier plotting
df['time_minutes'] = df['time'].dt.total_seconds() / 60
```

```python
# --- BASIC SUMMARY ---
print("Basic Info:")
print(df.info())
print("\nSummary Statistics:")
print(df.describe(include='all'))
```

```python
df.columns
```

```python
# Create a Dataframe containing unique rows for each runner
runners = df[['bib_number', 'name', 'country', 'city_state', 'age_group', 'gender']]
runners = runners.drop_duplicates(subset='bib_number', keep='first')
runners = runners.reset_index(drop=True)
print("\nUnique Runners DataFrame:", runners.shape)
```

```python
# --- GENDER DISTRIBUTION ---
plt.figure(figsize=(6,4))
sns.countplot(data=runners, x='gender', palette='pastel')
plt.title('Participant Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# --- AGE GROUP DISTRIBUTION ---
plt.figure(figsize=(10,5))
sns.countplot(data=runners, x='age_group', order=sorted(df['age_group'].dropna().unique()), palette='muted')
plt.title('Participant Count by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# --- FINISH TIME DISTRIBUTION ---
plt.figure(figsize=(10,5))
temp1 = df[['bib_number', 'time_minutes']].drop_duplicates(subset='bib_number', keep='last')
sns.histplot(temp1['time_minutes'].dropna(), bins=50, kde=True)
plt.title('Distribution of Finish Times (in minutes)')
plt.xlabel('Finish Time (minutes)')
plt.ylabel('Number of Runners')
plt.show()
```

```python
# --- BOX PLOT: FINISH TIME BY GENDER ---
plt.figure(figsize=(8,6))
temp2 = df[['bib_number', 'gender', 'time_minutes']].drop_duplicates(subset='bib_number', keep='last')
sns.boxplot(data=temp2, x='gender', y='time_minutes')
plt.title('Finish Time by Gender')
plt.ylabel('Time (minutes)')
plt.xlabel('Gender')
plt.show()
```

```python
# --- MISSING VALUES ---
missing = df.isnull().mean().sort_values(ascending=False)
missing = missing[missing > 0]
print("Missing Values (%):")
print(missing * 100)
```

```python
import plotly.express as px

fig = px.histogram(temp2, x='time_minutes', nbins=100,
                   title='Interactive Finish Time Distribution',
                   labels={'time_minutes': 'Finish Time (minutes)'},
                   color='gender')
fig.update_layout(bargap=0.1)
# fig.show()
fig.write_html("../plotly_charts/avg_pace_by_gender.html")
```

```python
# Make a copy and drop NaNs
split_df = df[df['split'].notna()].copy()

# Define a mapping for split labels
split_km_map = {
    '5K': 5,
    '10K': 10,
    '15K': 15,
    '20K': 20,
    'HALF': 21.1,
    '25K': 25,
    '30K': 30,
    '35K': 35,
    '40K': 40,
    'FINISH': 42.2
}

# Map values to new column
split_df['split_km'] = split_df['split'].map(split_km_map)
```

```python
avg_pace = split_df.groupby(['split_km', 'gender'])['min/km'].mean().reset_index()

fig = px.line(avg_pace, x='split_km', y='min/km',
              color='gender', markers=True,
              title='Average Pace by Distance and Gender',
              labels={'split_km': 'Distance (km)', 'min/km': 'Average Pace (min/km)'})
# fig.show()
fig.write_html("../plotly_charts/avg_pace_by_distance_and_gender.html")
```
